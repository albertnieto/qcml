````python
def classic_perceptron():
    model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    return model
````

Sequential groups a linear stack of layers into a Model.

Dense implements the operation: 

```
output = activation(dot(input, kernel) + bias) 
```

where activation is the element-wise activation function passed as the activation argument, kernel is a weights matrix created by the layer, and bias is a bias vector created by the layer (only applicable if use_bias is True).

````python
# Build a simple Sequential model
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

# Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')

# Weight and bias
weight = model.layers[0].get_weights()[0]
bias = model.layers[0].get_weights()[1]

# Summary
model.summary()
````

````python
print(weight, bias)
````

````python
# Declare model inputs and outputs for training
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
````

````python
# Train the model
model.fit(xs, ys, epochs=500)
````

````python
# Make a prediction
print(model.predict([10.0]))
````

````python
# Weight and bias
weight = model.layers[0].get_weights()[0]
bias = model.layers[0].get_weights()[1]
print(weight, bias)
````

````python
def draw_result_in_latex(circuit):
    job = backend.run(circuit)
    result = job.result()
    sv = result.get_statevector()
    latex = sv.draw(output='latex')
    return latex
````

````python
def get_sv(circuit):
    job = backend.run(circuit)
    result = job.result()
    return result.get_statevector()
````

````python
ket_0 = np.array([1, 0])
ket_1 = np.array([0, 1])
````

````python
a1 = (ket_0 + (math.e**(((0+1j)*math.pi)/4)) * ket_1) / np.sqrt(2)
a1
````

````python
qc = QuantumCircuit(1)
qc.initialize(a1, qc.qubits)
draw_result_in_latex(qc)
````

````python
a2 = ((math.e**(((0-1j)*math.pi)/4)) * ket_0 + ket_1) / np.sqrt(2)
a2
````

````python
qc = QuantumCircuit(1)
qc.initialize(a2, qc.qubits)
draw_result_in_latex(qc)
````
